```python
import time
import serial
import numpy as np
from numpy.random import randn,rand
import pyautogui, sys

import random
```

```python
>>> dev = serial.Serial("/dev/ttyACM0", baudrate=9600)
```

```python
# mouseposition Array

test = np.round(rand(6,15),3)
print(test)

for i in np.arange(test.shape[1]):
    j = test[:,i]
    #j = np.concatenate([["H"],j,["e"]])
    j = np.concatenate([["H"],j])
    print(j)
    for i in j:
        print(i)
        dev.write(i.encode())
        dev.write(b',')
    time.sleep(3)

```

```python
# mouseposition Karte
try:
    while True:
        x,y = pyautogui.position()
        liste = ['H',x,y]
        #print(liste)
        for i in liste:
            #print(i)
            dev.write(str(i).encode())
            dev.write(b',')
except KeyboardInterrupt:
    pass
```
